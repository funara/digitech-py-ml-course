## 2. Variabel & Tipe Data <a id='2'></a>

**Variabel** adalah tempat untuk menyimpan data.  
Di Python, kamu tidak perlu mendeklarasikan tipe data — Python mendeteksinya secara otomatis.

| Tipe Data | Contoh | Keterangan |
|-----------|--------|------------|
| `int` | `10`, `-3` | Bilangan bulat |
| `float` | `3.14`, `-0.5` | Bilangan desimal |
| `str` | `"Halo"`, `'Python'` | Teks / string |
| `bool` | `True`, `False` | Benar atau salah |

````python
nama = "Andi"          # str
umur = 20              # int
nilai = 87.5           # float
lulus = True           # bool

print(nama)
print(umur)
print(nilai)
print(lulus)
````

````python
# Fungsi type() digunakan untuk mengecek tipe data sebuah variabel
print(type(nama))
print(type(umur))
print(type(nilai))
print(type(lulus))
````

## 3. Operasi Aritmatika <a id='3'></a>

Python mendukung semua operasi matematika dasar:

| Operator | Fungsi | Contoh |
|----------|--------|--------|
| `+` | Penjumlahan | `5 + 3 = 8` |
| `-` | Pengurangan | `5 - 3 = 2` |
| `*` | Perkalian | `5 * 3 = 15` |
| `/` | Pembagian | `5 / 2 = 2.5` |
| `//` | Pembagian bulat | `5 // 2 = 2` |
| `%` | Sisa bagi (modulo) | `5 % 2 = 1` |
| `**` | Pangkat | `5 ** 2 = 25` |

````python
a = 10
b = 3

print("Penjumlahan:", a + b)
print("Pengurangan:", a - b)
print("Perkalian:", a * b)
print("Pembagian:", a / b)
print("Pembagian Bulat:", a // b)
print("Sisa Bagi:", a % b)
print("Pangkat:", a ** b)
````

## 4. String (Teks) <a id='4'></a>

String adalah tipe data untuk menyimpan teks.  
Ada banyak hal yang bisa kamu lakukan dengan string di Python.

````python
kalimat = "Belajar Python itu menyenangkan!"

print(len(kalimat))          # Panjang string
print(kalimat.upper())       # Huruf besar semua
print(kalimat.lower())       # Huruf kecil semua
print(kalimat.replace("menyenangkan", "mudah"))  # Ganti kata
````

````python
# f-string: cara modern untuk memasukkan variabel ke dalam string
nama = "Siti"
ipk = 3.87

pesan = f"Mahasiswa {nama} memiliki IPK {ipk}"
print(pesan)
````

````python
# Mengakses karakter dengan index (dimulai dari 0)
kata = "Python"
print(kata[0])   # Karakter pertama: P
print(kata[-1])  # Karakter terakhir: n
print(kata[0:3]) # Karakter ke-0 sampai ke-2: Pyt
````

## 6. Kondisi (if / elif / else) <a id='6'></a>

Kondisi digunakan untuk membuat program yang bisa **mengambil keputusan**.

```
if kondisi_1:
    # lakukan ini jika kondisi_1 benar
elif kondisi_2:
    # lakukan ini jika kondisi_2 benar
else:
    # lakukan ini jika semua kondisi salah
```

Operator perbandingan: `==`, `!=`, `>`, `<`, `>=`, `<=`

````python
nilai = 75

if nilai >= 80:
    print("Grade: A")
elif nilai >= 70:
    print("Grade: B")
elif nilai >= 60:
    print("Grade: C")
else:
    print("Grade: D — perlu belajar lebih keras!")
````

````python
# Operator logika: and, or, not
umur = 20
punya_ktp = True

if umur >= 17 and punya_ktp:
    print("Boleh memilih dalam pemilu.")
else:
    print("Belum boleh memilih.")
````

## 7. Perulangan (for & while) <a id='7'></a>

Perulangan digunakan untuk menjalankan blok kode **berulang kali**.

**`for`** — digunakan saat kamu tahu berapa kali perulangan dilakukan.  
**`while`** — digunakan selama suatu kondisi masih `True`.

````python
# for loop dengan range()
# range(5) menghasilkan: 0, 1, 2, 3, 4
for i in range(5):
    print(f"Iterasi ke-{i}")
````

````python
# for loop pada list
buah = ["apel", "mangga", "pisang", "jeruk"]

for b in buah:
    print(f"Saya suka {b}")
````

````python
# while loop
hitung = 1
while hitung <= 5:
    print(f"Hitungan: {hitung}")
    hitung += 1  # Sama dengan: hitung = hitung + 1
````

## 8. Fungsi <a id='8'></a>

**Fungsi** adalah blok kode yang bisa dipanggil berulang kali.  
Fungsi membantu kode menjadi lebih **terorganisir dan tidak berulang**.

```python
def nama_fungsi(parameter):
    # isi fungsi
    return hasil
```

````python
# Fungsi sederhana tanpa parameter
def sapa():
    print("Selamat datang di mata kuliah Machine Learning!")

sapa()  # Memanggil fungsi
````

````python
# Fungsi dengan parameter dan return
def hitung_luas_lingkaran(jari_jari):
    pi = 3.14159
    luas = pi * jari_jari ** 2
    return luas

hasil = hitung_luas_lingkaran(7)
print(f"Luas lingkaran dengan r=7: {hasil:.2f}")
````

````python
# Fungsi dengan beberapa parameter
def tentukan_grade(nilai, kehadiran):
    if kehadiran < 75:
        return "Tidak Lulus (kehadiran)"
    elif nilai >= 80:
        return "A"
    elif nilai >= 70:
        return "B"
    elif nilai >= 60:
        return "C"
    else:
        return "D"

print(tentukan_grade(85, 80))
print(tentukan_grade(75, 70))
print(tentukan_grade(55, 90))
````
