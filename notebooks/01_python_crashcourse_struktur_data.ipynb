## Apa itu Struktur Data?

Struktur data adalah cara **menyimpan dan mengorganisir banyak data** sekaligus.  
Kalau variabel biasa hanya menyimpan 1 nilai, struktur data bisa menyimpan **puluhan, ratusan, bahkan ribuan** nilai.

Di notebook ini kita akan belajar 4 struktur data utama Python:

| Struktur | Simbol | Sifat | Contoh Penggunaan |
|----------|--------|-------|-------------------|
| **List** | `[ ]` | Berurutan, bisa diubah | Daftar nilai mahasiswa |
| **Tuple** | `( )` | Berurutan, tidak bisa diubah | Koordinat GPS |
| **Dictionary** | `{ }` | Pasangan key-value | Data profil pengguna |
| **Set** | `{ }` | Tidak berurutan, unik | Daftar kategori unik |

## 1. List

List adalah struktur data paling sering digunakan.  
Bayangkan list seperti **laci lemari** — tiap laci (index) menyimpan satu nilai, dimulai dari index `0`.

```
nilai = [85, 70, 90, 65, 80]
index:    0   1   2   3   4
```

````python
# Membuat list
nilai_ujian = [85, 70, 90, 65, 80]
nama_mahasiswa = ["Andi", "Budi", "Cici", "Dedi"]
campuran = [1, "hello", 3.14, True]  # List bisa menyimpan tipe berbeda

print(nilai_ujian)
print(nama_mahasiswa)
````

````python
# Mengakses elemen
print(nilai_ujian[0])    # Elemen pertama
print(nilai_ujian[-1])   # Elemen terakhir
print(nilai_ujian[1:4])  # Elemen index 1 sampai 3 (slicing)
````

````python
# Operasi pada list
buah = ["apel", "mangga"]

buah.append("pisang")    # Tambah di akhir
print("Setelah append:", buah)

buah.insert(1, "jeruk")  # Tambah di index tertentu
print("Setelah insert:", buah)

buah.remove("mangga")    # Hapus berdasarkan nilai
print("Setelah remove:", buah)

print("Panjang list:", len(buah))
````

````python
# List dengan fungsi bawaan
angka = [5, 2, 8, 1, 9, 3]

print("Min:", min(angka))
print("Max:", max(angka))
print("Sum:", sum(angka))
print("Sorted:", sorted(angka))
````

````python
# Iterasi list dengan for
nama_mahasiswa = ["Andi", "Budi", "Cici", "Dedi"]

for nama in nama_mahasiswa:
    print(f"Selamat datang, {nama}!")
````

````python
# enumerate() — iterasi sekaligus dapat indexnya
for index, nama in enumerate(nama_mahasiswa):
    print(f"{index + 1}. {nama}")
````

## 2. Tuple

Tuple mirip list, tapi **tidak bisa diubah** setelah dibuat (immutable).  
Gunakan tuple untuk data yang **tidak boleh berubah**, seperti koordinat, tanggal, atau konfigurasi.

````python
koordinat_jakarta = (-6.2088, 106.8456)  # latitude, longitude

print("Koordinat:", koordinat_jakarta)
print("Latitude:", koordinat_jakarta[0])
print("Longitude:", koordinat_jakarta[1])
````

````python
# Tuple tidak bisa diubah — ini akan error!
try:
    koordinat_jakarta[0] = -7.0
except TypeError as e:
    print(f"Error: {e}")
````

````python
# Tuple packing & unpacking
data_mahasiswa = ("Andi", 20, 3.85)  # nama, umur, ipk

# Unpacking — ambil nilai ke variabel terpisah
nama, umur, ipk = data_mahasiswa
print(f"Nama: {nama}, Umur: {umur}, IPK: {ipk}")
````

## 3. Dictionary

Dictionary menyimpan data dalam pasangan **key (kunci) : value (nilai)**.  
Bayangkan seperti **kamus** — kamu cari kata (key), lalu dapat definisinya (value).

```python
profil = {
    "nama": "Andi",
    "umur": 20
}
#           ↑ key    ↑ value
```

````python
# Membuat dictionary
mahasiswa = {
    "nama": "Budi Santoso",
    "nim": "20230001",
    "ipk": 3.75,
    "aktif": True
}

# Mengakses nilai
print(mahasiswa["nama"])
print(mahasiswa["ipk"])
````

````python
# Menambah dan mengubah data
mahasiswa["jurusan"] = "Informatika"    # Tambah key baru
mahasiswa["ipk"] = 3.80                 # Ubah nilai yang ada

print(mahasiswa)
````

````python
# Mengecek apakah key ada
if "nama" in mahasiswa:
    print("Key 'nama' ada:", mahasiswa["nama"])

# .get() — aman karena tidak error jika key tidak ada
print(mahasiswa.get("alamat", "Alamat tidak tersedia"))
````

````python
# Iterasi dictionary
for key, value in mahasiswa.items():
    print(f"{key}: {value}")
````

````python
# List of dictionaries — sangat umum digunakan!
# Ini mirip dengan tabel di spreadsheet
dataset = [
    {"nama": "Andi", "nilai": 85, "kelas": "A"},
    {"nama": "Budi", "nilai": 70, "kelas": "B"},
    {"nama": "Cici", "nilai": 92, "kelas": "A"},
]

for mahasiswa in dataset:
    print(f"{mahasiswa['nama']} - Nilai: {mahasiswa['nilai']}")
````

## 4. Set

Set adalah koleksi yang **tidak memiliki duplikat** dan **tidak berurutan**.  
Sangat berguna untuk mencari nilai unik dari sebuah data.

````python
# Set otomatis menghapus duplikat
data_kota = ["Jakarta", "Surabaya", "Jakarta", "Bandung", "Surabaya", "Medan"]

kota_unik = set(data_kota)
print("Data asli:", data_kota)
print("Nilai unik:", kota_unik)
print("Jumlah kota unik:", len(kota_unik))
````

````python
# Operasi set — mirip operasi himpunan di matematika
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print("Gabungan (union):", set_a | set_b)
print("Irisan (intersection):", set_a & set_b)
print("Selisih A-B:", set_a - set_b)
````

## 5. List Comprehension (Bonus — Cara Elegan Python)

List comprehension adalah cara **membuat list baru dari list lama** dalam satu baris.  
Ini adalah fitur Python yang sangat populer karena ringkas dan mudah dibaca.

````python
# Cara biasa — membuat list kuadrat
kuadrat = []
for i in range(1, 6):
    kuadrat.append(i ** 2)
print("Cara biasa:", kuadrat)
````

````python
# Cara list comprehension — lebih ringkas!
kuadrat = [i ** 2 for i in range(1, 6)]
print("List comprehension:", kuadrat)
````

````python
# List comprehension dengan kondisi
nilai = [85, 55, 90, 45, 70, 60, 80]

# Ambil hanya nilai yang >= 70
nilai_lulus = [n for n in nilai if n >= 70]
print("Yang lulus:", nilai_lulus)
````
